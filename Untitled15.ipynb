```python
import os
import json
import pandas as pd
import matplotlib.pyplot as plt

# === PATHS ===
BASE_DIR = "/content/dataset_ai_1_unzipped/dataset_ai_1"
IMG_DIR = os.path.join(BASE_DIR, "bin-images")
META_DIR = os.path.join(BASE_DIR, "metadata")

# === FIND MATCHING FILES ===
img_files = set(f.replace(".jpg", "") for f in os.listdir(IMG_DIR) if f.endswith(".jpg"))
meta_files = set(f.replace(".json", "") for f in os.listdir(META_DIR) if f.endswith(".json"))

# Keep only common ones
common_files = sorted(list(img_files.intersection(meta_files)))

# Print summary
print(f"‚úÖ Matching image‚Äìmetadata pairs: {len(common_files)}")
print(f"üñºÔ∏è Extra images: {len(img_files - meta_files)}")
print(f"üìÑ Extra metadata: {len(meta_files - img_files)}")

# === SAFE GETTER FUNCTION ===
def safe_get_value(details, key):
    """Safely extract nested 'value' field from metadata."""
    if key in details and isinstance(details[key], dict):
        return details[key].get("value", None)
    return None

# === BUILD ITEM-LEVEL DATAFRAME ===
records = []
for file_id in common_files:
    json_path = os.path.join(META_DIR, f"{file_id}.json")
    with open(json_path, "r") as f:
        data = json.load(f)

    bin_data = data.get("BIN_FCSKU_DATA", {})
    total_qty = 0

    for asin, details in bin_data.items():
        name = details.get("name", "")
        qty = details.get("quantity", 0)
        total_qty += qty

        records.append({
            "file_id": file_id,
            "asin": asin,
            "name": name,
            "quantity": qty,
            "height": safe_get_value(details, "height"),
            "length": safe_get_value(details, "length"),
            "width": safe_get_value(details, "width"),
            "weight": safe_get_value(details, "weight")
        })

df = pd.DataFrame(records)
print("\nüì¶ Item-level DataFrame:")
print(df.head())

# === CHECK MISSING VALUES ===
print("\nüîç Missing values per column:")
print(df[["height", "length", "width", "weight"]].isna().sum())

# === BIN-LEVEL SUMMARY ===
bin_summary = (
    df.groupby("file_id")
      .agg(num_items=("asin", "count"),
           total_quantity=("quantity", "sum"))
      .reset_index()
)

print("\nüß∫ Bin Summary (first 5 rows):")
print(bin_summary.head())

# === DISTRIBUTIONS ===
plt.figure(figsize=(10,4))
plt.hist(bin_summary['num_items'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution: Number of Items per Bin')
plt.xlabel('Items per bin')
plt.ylabel('Count')
plt.show()

plt.figure(figsize=(10,4))
plt.hist(bin_summary['total_quantity'], bins=20, color='salmon', edgecolor='black')
plt.title('Distribution: Total Quantity per Bin')
plt.xlabel('Total quantity')
plt.ylabel('Count')
plt.show()

# === MOST FREQUENT ITEMS ===
top_items = (
    df.groupby("name")["file_id"]
      .count()
      .sort_values(ascending=False)
      .head(10)
)

plt.figure(figsize=(10,6))
top_items.plot(kind='barh', color='teal')
plt.title("Top 10 Most Frequent Items in Bins")
plt.xlabel("Count")
plt.ylabel("Item Name")
plt.gca().invert_yaxis()
plt.show()

# === DIMENSION STATISTICS ===
dim_cols = ["height", "length", "width", "weight"]
print("\nüìè Dimension Statistics:")
print(df[dim_cols].describe())

# === VISUALIZE MISSING VALUE RATIOS ===
missing_ratio = df[dim_cols].isna().mean() * 100
plt.figure(figsize=(8,4))
missing_ratio.plot(kind='bar', color='orange', edgecolor='black')
plt.title("Percentage of Missing Values per Dimension Field")
plt.ylabel("Missing Value Percentage (%)")
plt.show()
```

```python
import seaborn as sns
import numpy as np

# Drop NaN for correlation computation
numeric_df = df[["height", "length", "width", "weight", "quantity"]].dropna()

plt.figure(figsize=(8,6))
corr = numeric_df.corr()
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap: Dimensions, Weight, and Quantity")
plt.show()
```

```python
import os
import json
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.image as mpimg

# === PATHS ===
BASE_DIR = "/content/dataset_ai_1_unzipped/dataset_ai_1"
IMG_DIR = os.path.join(BASE_DIR, "bin-images")
META_DIR = os.path.join(BASE_DIR, "metadata")

# === MATCHING FILES ===
img_files = set(f.replace(".jpg", "") for f in os.listdir(IMG_DIR) if f.endswith(".jpg"))
meta_files = set(f.replace(".json", "") for f in os.listdir(META_DIR) if f.endswith(".json"))
common_files = sorted(list(img_files.intersection(meta_files)))

print(f"‚úÖ Matching pairs: {len(common_files)}")
print(f"üñºÔ∏è Extra images: {len(img_files - meta_files)}")
print(f"üìÑ Extra metadata: {len(meta_files - img_files)}")

# === SAFE GETTER ===
def safe_get_value(details, key):
    if key in details and isinstance(details[key], dict):
        return details[key].get("value", None)
    return None

# === ITEM-LEVEL DATAFRAME ===
records = []
for file_id in common_files:
    json_path = os.path.join(META_DIR, f"{file_id}.json")
    with open(json_path, "r") as f:
        data = json.load(f)
    bin_data = data.get("BIN_FCSKU_DATA", {})

    for asin, details in bin_data.items():
        name = details.get("name", "")
        qty = details.get("quantity", 0)

        records.append({
            "file_id": file_id,
            "asin": asin,
            "name": name,
            "quantity": qty,
            "height": safe_get_value(details, "height"),
            "length": safe_get_value(details, "length"),
            "width": safe_get_value(details, "width"),
            "weight": safe_get_value(details, "weight")
        })

df = pd.DataFrame(records)
print("\nüì¶ Item-level DataFrame:")
print(df.head())

# === MISSING VALUE SUMMARY ===
print("\nüîç Missing values per column:")
print(df[["height", "length", "width", "weight"]].isna().sum())

# === BIN-LEVEL SUMMARY ===
bin_summary = df.groupby("file_id").agg(
    num_items=("asin", "count"),
    total_quantity=("quantity", "sum"),
    unique_items=("asin", "nunique")
).reset_index()

print("\nüß∫ Bin Summary (first 5 rows):")
print(bin_summary.head())

# === HISTOGRAMS: ITEMS AND QUANTITY PER BIN ===
plt.figure(figsize=(10,4))
plt.hist(bin_summary['num_items'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution: Number of Items per Bin')
plt.xlabel('Items per bin')
plt.ylabel('Count')
plt.show()

plt.figure(figsize=(10,4))
plt.hist(bin_summary['total_quantity'], bins=20, color='salmon', edgecolor='black')
plt.title('Distribution: Total Quantity per Bin')
plt.xlabel('Total quantity')
plt.ylabel('Count')
plt.show()

plt.figure(figsize=(10,4))
plt.hist(bin_summary['unique_items'], bins=20, color='lightgreen', edgecolor='black')
plt.title('Distribution: Unique Items per Bin')
plt.xlabel('Unique items')
plt.ylabel('Count')
plt.show()

# === MOST FREQUENT ITEMS ===
top_items = df.groupby("name")["file_id"].count().sort_values(ascending=False).head(10)
plt.figure(figsize=(10,6))
top_items.plot(kind='barh', color='teal')
plt.title("Top 10 Most Frequent Items in Bins")
plt.xlabel("Count")
plt.ylabel("Item Name")
plt.gca().invert_yaxis()
plt.show()

# === DIMENSION STATISTICS ===
dim_cols = ["height", "length", "width", "weight"]
print("\nüìè Dimension Statistics:")
print(df[dim_cols].describe())

# === MISSING VALUE RATIOS PLOT ===
missing_ratio = df[dim_cols].isna().mean() * 100
plt.figure(figsize=(8,4))
missing_ratio.plot(kind='bar', color='orange', edgecolor='black')
plt.title("Percentage of Missing Values per Dimension Field")
plt.ylabel("Missing Value Percentage (%)")
plt.show()

# === VOLUME VS WEIGHT ===
df["volume"] = df["height"] * df["length"] * df["width"]
plt.figure(figsize=(7,5))
plt.scatter(df["volume"], df["weight"], alpha=0.5)
plt.title("Item Volume vs Weight")
plt.xlabel("Volume (in¬≥)")
plt.ylabel("Weight (pounds)")
plt.show()
print("Correlation between volume and weight:", df[["volume", "weight"]].corr().iloc[0,1])

# === CORRELATION HEATMAP ===
numeric_df = df[["height", "length", "width", "weight", "quantity"]].dropna()
plt.figure(figsize=(8,6))
sns.heatmap(numeric_df.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap: Dimensions, Weight, Quantity")
plt.show()

# === BOX PLOTS FOR OUTLIERS ===
plt.figure(figsize=(8,5))
sns.boxplot(data=df[["height", "length", "width", "weight"]], orient="h")
plt.title("Boxplot of Item Dimensions and Weight")
plt.show()

# === TOP ITEMS WITH MISSING METADATA ===
missing_by_item = df.groupby("name")[dim_cols].apply(lambda x: x.isna().mean() * 100).mean(axis=1).sort_values(ascending=False).head(10)
plt.figure(figsize=(10,6))
missing_by_item.plot(kind='barh', color='tomato')
plt.title("Top 10 Items with Most Missing Metadata (%)")
plt.xlabel("Missing Percentage")
plt.ylabel("Item Name")
plt.gca().invert_yaxis()
plt.show()

# === SAMPLE IMAGE CHECK ===
sample_id = df["file_id"].sample(1).values[0]
img_path = os.path.join(IMG_DIR, f"{sample_id}.jpg")
meta_sample = df[df["file_id"] == sample_id]

print(f"üñºÔ∏è Sample Bin ID: {sample_id}")
display(meta_sample[["asin", "name", "quantity"]])

img = mpimg.imread(img_path)
plt.imshow(img)
plt.axis("off")
plt.title(f"Sample Bin Image: {sample_id}")
plt.show()
```
